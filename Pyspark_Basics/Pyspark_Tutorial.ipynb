**Adding Columns to Existing Dataframe**

```python
# Adding a column
df = df.withColumn("half acidity",df["fixed acidity"]*0.5)
```

```python
df.show()
```

```python
## Rename the column
df.withColumnRenamed("fixed acidity","acidity fixed").show()
```

#### 3rd Section

**In 3rd Section of the tutorial we look at Different Filtering Options of the data frame**

```python
### Filter basis the column and conditions
df.filter("density<=0.9").show()
```

```python
df.filter("density<=30").select("volatile acidity", "alcohol").show()
```

```python
## Taking Multiple Condition (And Operation)
df.filter((df["density"]<=30) & (df["alcohol"]<9.2)).show()
```

```python
# Not Operation
df.filter(~(df["alcohol"]<=9.2)).show()
```
