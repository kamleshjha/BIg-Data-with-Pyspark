#### Basic Operations with Spark Dataframes

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('bank').getOrCreate()
df = spark.read.csv('/content/drive/MyDrive/Colab Notebooks/data/bank_data.csv',header=True,inferSchema=True)
df.show(10)
```

```python
df.printSchema()
```

```python
df.filter("balance < 500").show()
```

```python
df.filter("balance < 500").select(['marital','Education']).show()
```

```python
df.filter((df['marital'] == 'single') & (df['Education'] == 'secondary')).show()
```

```python
### Its not the Second Condition

df.filter((df['marital'] == 'single') & ~(df['Education'] == 'secondary')).show()
```

```python
### Saving Results in a separate Dataframe using Collect() function

df.filter(df['marital'] == 'single').show(4)
```

```python
results = df.filter(df['marital'] == 'single').collect()
```

```python
## Accessing Results values from the saved collection using Dictionary
row = results[0]
row.asDict()['balance']
```

**Bonus Question!**
There are too many decimal places for mean and stddev in the describe() dataframe. Format the numbers to just show up to two decimal places. Pay careful attention to the datatypes that .describe() returns, we didn't cover how to do this exact formatting, but we covered something very similar. Check this link for a hint

```python
from pyspark.sql.functions import format_number
new_df = walmart.describe()
new_df.select(new_df['summary'],
              format_number(new_df['Open'].cast('float'),2).alias('Open'),
              format_number(new_df['High'].cast('float'),2).alias('High'),
              format_number(new_df['Low'].cast('float'),2).alias('Low'),
              format_number(new_df['Close'].cast('float'),2).alias('Close'),
              new_df['Volume'].cast('int').alias('Vol'),
              format_number(new_df['Adj Close'].cast('float'),2).alias('AdjClose')
              ).show()
```

Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day

```python
walmart_df = walmart.withColumn("HV Ratio",walmart["High"]/walmart["Volume"])
walmart_df.select('HV Ratio').show(10)
```

What day had the Peak High in Price?

```python
walmart_df.orderBy(walmart_df["High"].desc()).head(1)[0][0]
```

Mean of Close Column ?

```python
from pyspark.sql.functions import mean
walmart_df.select(mean("Close")).show()
```

What is Max and Min of Volume Column

```python
from pyspark.sql.functions import max,min
walmart_df.select(max("Volume"),min("Volume")).show()
```

how Many days was close lower than 60 dollars

```python
walmart_df.filter(walmart_df["Close"] < 60).count()
```

What Percentage of time was High greater than 80 dollars

```python
(walmart_df.filter(walmart_df["High"]>80).count()*1.0/walmart_df.count())*100
```

Implement Pearson Correlation Coefficient between High and Volume Columns in Dataset

```python
from pyspark.sql.functions import corr
walmart_df.select(corr("High","Volume").alias("Pearson Correlation between High and Volume")).show()
```

What is the max High per year?

```python
from pyspark.sql.functions import year
yeardf = walmart_df.withColumn("Year",year(walmart_df["Date"]))
max_df = yeardf.groupBy('Year').max()
max_df.select('Year','max(High)').show()
```

What is the average Close for each Calendar Month?
In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months

```python
from pyspark.sql.functions import month
monthdf = walmart_df.withColumn("Month",month("Date"))
monthavgs = monthdf.select("Month","Close").groupBy("Month").mean()
monthavgs.select("Month","avg(Close)").orderBy('Month').show()
```
